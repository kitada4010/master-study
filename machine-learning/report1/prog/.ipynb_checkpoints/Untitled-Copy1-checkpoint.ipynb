```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Num = 10
x = np.random.rand(Num)
b = np.random.rand(1)
print(b)
y = b*x + np.random.rand(Num)*0.5
xx = np.linspace(min(x), max(x), Num)
plt.plot(x,y,marker='.',ls='')
#plt.plot(y)
#plt.plot(xx)
plt.show()
```
